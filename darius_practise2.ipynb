```python
train = pd.read_csv('data/train.csv')
```

```python
df = train[:1000]
```

```python
price_category_mean = df[['price', 'category_name']].groupby('category_name').mean().reset_index()
price_category_std = df[['price', 'category_name']].groupby('category_name').std().reset_index()
```

```python
for x in range (0,len(df)-1):
    df.norm_price[x] =  df.price[x] - price_category_mean.price [ price_category_mean.category_name == df.category_name[x]] 
    df.norm_price[x] =  df.norm_price[x] / price_category_std.price [ price_category_std.category_name == df.category_name[x]]
df
```
